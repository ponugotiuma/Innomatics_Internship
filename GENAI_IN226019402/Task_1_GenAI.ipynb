## **Step 1: Import Libraries**

```python
from langchain_core.prompts import PromptTemplate
from langchain_core.prompts import ChatPromptTemplate
```

## **Task 1: Replace Hardcoded Prompts**

In this task, I converted a hardcoded prompt into a reusable format using LangChain’s PromptTemplate. Earlier, the prompt was written using an f-string, which works but is not flexible. By using PromptTemplate, I can now reuse the same structure for different topics without rewriting the code. This makes the prompt system cleaner and more scalable.

```python
# Template
template = "Explain {topic} in simple terms for beginners"

prompt_template = PromptTemplate(
    input_variables=["topic"],
    template=template
)

print(prompt_template.format(topic="Machine Learning"))
```

## **Task 2: Multi-Input Prompt System**

Here, I extended the prompt to take multiple inputs like topic, audience, and tone. This makes the prompt more dynamic and useful in real situations because different users may need different styles of explanation. For example, explaining AI to beginners is very different from explaining it to engineers. This task helped me understand how to make prompts more customizable.

```python
multi_template = "Explain {topic} for {audience} in a {tone} tone"

multi_prompt = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template=multi_template
)

print(multi_prompt.format(topic="AI", audience="beginners", tone="friendly"))
print(multi_prompt.format(topic="Python", audience="kids", tone="fun"))
print(multi_prompt.format(topic="Deep Learning", audience="engineers", tone="technical"))
```

## **Task 3: Prompt Variations Engine**

In this task, I created different types of prompts for different purposes such as teaching, interview, and storytelling. This shows how changing the prompt structure can change the type of output we get from the model. For example, one prompt explains step by step, another asks questions, and another tells a story. This helped me understand the importance of prompt design.

```python
# Template 1
teaching_prompt = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

# Template 2
interview_prompt = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 interview questions about {topic}"
)

# Template 3
story_prompt = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)

topic = "Machine Learning"

print(teaching_prompt.format(topic=topic))
print(interview_prompt.format(topic=topic))
print(story_prompt.format(topic=topic))
```

## **Task 4: ChatPromptTemplate System**

This task introduces chat-based prompts using roles like teacher, interviewer, and motivator. Instead of just giving a simple instruction, we define how the AI should behave. The system message sets the role, and the user message gives the input. This is useful for building chatbots or assistants where the response style matters.

```python
def chat_prompt(role, topic):

    if role == "teacher":
        system_message = "You are a helpful teacher."
    elif role == "interviewer":
        system_message = "You are a strict interviewer."
    elif role == "motivator":
        system_message = "You are an inspiring motivator."
    else:
        system_message = "You are a helpful assistant."

    chat_template = ChatPromptTemplate.from_messages([
        ("system", system_message),
        ("user", "Explain {topic}")
    ])

    return chat_template.format_messages(topic=topic)

# 1
messages = chat_prompt("teacher", "Neural Networks")
print(messages)
# 2
messages = chat_prompt("interviewer", "General Science")
print(messages)
# 3
messages = chat_prompt("motivator", "Life Lesson")
print(messages)
# 4
messages = chat_prompt("student", "Machine Learning")
print(messages)
```

## **Task 5: Input Validation Layer**

In this task, I added validation to check whether the inputs like audience and tone are valid or not. This is important because users might enter incorrect values. If the input is invalid, I either assign a default value or handle it properly. This makes the system more reliable and prevents unexpected results.

```python
def validate_inputs(audience, tone):

    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        print("Invalid audience! Defaulting to 'beginner'")
        audience = "beginner"

    if tone not in valid_tone:
        print("Invalid tone! Defaulting to 'formal'")
        tone = "formal"

    return audience, tone

audience, tone = validate_inputs("kids", "friendly")
print(audience, tone)
print("-------------------------------------------")
audience, tone = validate_inputs("expert", "formal")
print(audience, tone)
```

## **Task 6: Prompt Generator App**

Here, I combined everything into one function that generates prompts based on user input. It takes topic, audience, tone, and style, validates the inputs, and then selects the correct template. This feels like a small real-world application where prompts are generated dynamically. It shows how all the concepts can work together.

```python
def generate_prompt(topic, audience, tone, style):

    audience, tone = validate_inputs(audience, tone)

    if style == "teaching":
        template = "Explain {topic} for {audience} in a {tone} teaching style"

    elif style == "interview":
        template = "Ask interview questions about {topic} for {audience} in a {tone} tone"

    elif style == "storytelling":
        template = "Explain {topic} for {audience} in a {tone} storytelling style"

    else:
        template = "Explain {topic} for {audience} in a {tone} tone"

    prompt = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=template
    )

    return prompt.format(topic=topic, audience=audience, tone=tone)

print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

## **Task 7: Template Reusability Test**

In this task, I used a single template and tested it with multiple topics. The goal was to show that one template can be reused without changing its structure. This reduces code repetition and keeps things simple. It also proves that PromptTemplate is efficient for handling multiple inputs.

```python
reusable_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms"
)

topics = ["AI", "Blockchain", "Cloud Computing", "Cyber Security", "Data Science"]

for topic in topics:
    print(reusable_template.format(topic=topic))
```

Overall, the system works in a step-by-step flow. First, the user gives input. Then the input is validated to make sure it is correct. After that, the appropriate prompt template is selected, and a dynamic prompt is generated. This structured flow makes the system organized and easy to scale in real applications.
