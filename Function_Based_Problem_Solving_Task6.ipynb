```python
# 1.SMART PARKING LOT MANAGEMENT SYSTEM

def parking_system(capacity, logs):
    parked = 0
    peak = 0

    for log in logs:
        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1
        peak = max(peak, parked)

    print("Currently Parked Vehicles:", parked)
    if parked > capacity:
        print("Parking Status: Full")
    else:
        print("Parking Status: Available")

parking_system(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

```python
# 2.ONLINE FOOD DELIVERY TIME ESTIMATOR

def delivery_eta(distance, traffic, weather):
    time = distance * 5   # base: 5 min per km

    if traffic.lower() == "high":
        time += 15
    elif traffic.lower() == "medium":
        time += 8

    if weather.lower() == "rainy":
        time += 10
    elif weather.lower() == "stormy":
        time += 20

    print("Estimated Delivery Time:", time, "minutes")

delivery_eta(8, "High", "Rainy")
```

```python
# 3.MOVIE THEATRE SEAT OCCUPANCY ANALYZER

def theatre_occupancy(total_seats, booked_list):
    booked = len(booked_list)
    percent = (booked / total_seats) * 100

    print("Occupancy:", int(percent), "%")

    if percent == 100:
        print("Show Status: Housefull")
    elif percent >= 70:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Available")

theatre_occupancy(200, [1]*150)
```

```python
# 4.CLOUD SERVER LOAD CLASSIFICATION SYSTEM

def server_load(cpu_readings):
    avg = sum(cpu_readings) / len(cpu_readings)
    print("Average CPU Load:", int(avg), "%")

    if avg < 50:
        print("Server Status: Normal")
    elif avg <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")

server_load([45, 60, 70, 85, 90])
```

```python
# 5.SMART CLASSROOM RESOURCE USAGE MONITOR 

def resource_monitor(usage):
    overused = []

    for resource, hours in usage.items():
        if hours > 8:
            overused.append(resource)

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")

resource_monitor({
    "Projector": 6,
    "AC": 9,
    "Lights": 4
})
```

```python
# 6.ONLINE EVENT REGISTRATION CAPACITY CONTROLLER

def registration_control(capacity, registrations):
    confirmed = min(capacity, registrations)
    waitlisted = max(0, registrations - capacity)

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)

    if registrations >= capacity:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")

registration_control(100, 105)
```
