# Task-2 : Preprocessing Function

```python
import re

def preprocess_text(text):

    if not text:
        return [], ""

    text = text.lower()

    text = re.sub(r'http\S+|www\S+|\S+@\S+', '', text)

    text = re.sub(r'\d+', '', text)

    text = re.sub(r'[^a-z\s]', '', text)

    text = re.sub(r'(.)\1{2,}', r'\1\1', text)

    text = re.sub(r'\s+', ' ', text).strip()

    tokens = text.split()

    tokens = [
        word for word in tokens
        if len(word) > 2 or word in ['no', 'not']
    ]

    cleaned_sentence = " ".join(tokens)

    return tokens, cleaned_sentence

text = "WOW!!! This is GREAT!!! I have 2 dogs and sooo happy 😊😊"
tokens, cleaned = preprocess_text(text)
print("Original:",text)
print("Tokens:", tokens)
print("Cleaned Sentence:", cleaned)
```

# Task-3 : Stress Testing

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results = []

for text in test_sentences:
    tokens, cleaned = preprocess_text(text)

    print("\nOriginal:", text)
    print("Tokens:", tokens)
    print("Cleaned:", cleaned)

    results.append(tokens)
```

# Task-4 : Token Analysis

```python
analytics = []

for tokens in results:

    total = len(tokens)
    unique = len(set(tokens))
    avg_len = sum(len(w) for w in tokens) / total if total > 0 else 0

    analytics.append({
        "total_tokens": total,
        "unique_tokens": unique,
        "avg_token_length": round(avg_len, 2)
    })

print("\nAnalytics:")
for i, a in enumerate(analytics):
    print(f"Sentence {i+1}:", a)
```

# Task-5 : Frequency Analysis

```python
from collections import Counter

all_tokens = []

for tokens in results:
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("\nTop 10 frequent words:")
print(counter.most_common(10))

print("\nTop 5 least frequent words:")
print(counter.most_common()[:-6:-1])
```

# Task-6 : Full Pipeline

```python
def full_pipeline(text_list):

    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, cleaned = preprocess_text(text)

        all_tokens.extend(tokens)
        clean_sentences.append(cleaned)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10"
]
output = full_pipeline(test_sentences)

print(output)
```
