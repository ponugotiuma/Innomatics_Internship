```python
# 1.SOCIAL MEDIA -- POST ENGAGEMENT ANALYZER

def post_engagement_analyzer(likes_list):

    # Calculate total likes
    total_likes = sum(likes_list)
    
    # Determine post status
    if total_likes >= 1000:
        post_status = "Viral Post"
    else:
        post_status = "Normal Engagement"
    
    print(f"Total Likes: {total_likes}")
    print(f"Post Status: {post_status}")

likes_data = [200, 300, 250, 300] 
post_engagement_analyzer(likes_data)
```

```python
# 2.HEALTHCARE -- MEDICINE STOCK ALERT SYSTEM

def medicine_stock_alert(medicine, stock):
    if stock < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"
    print(f"Medicine: {medicine}")
    print(f"Stock: {stock}")
    print(f"Status: {status}")

medicine_stock_alert("Paracetamol", 6)
```

```python
# 3.AGRICULTURE -- RAINFALL ADEQUECY CHECKER

def rainfall_adequacy_checker(rainfall_data, required_level):
    
    average_rainfall = sum(rainfall_data) / len(rainfall_data)
    
    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    
    return average_rainfall, status

rainfall_data = [60, 80, 70, 78]
required_level = 70
avg_rainfall, rainfall_status = rainfall_adequacy_checker(rainfall_data, required_level)

print(f"Average Rainfall: {avg_rainfall:.0f}")
print(f"Rainfall Status: {rainfall_status}")
```

```python
# 4.SOCIAL MEDIA -- DUPLICATE ACCOUNT DETECTION

def detect_duplicate_usernames(usernames):
    seen = set()
    duplicates = set()
    for username in usernames:
        if username in seen:
            duplicates.add(username)
        else:
            seen.add(username)
    duplicate_status = "Duplicate Accounts Found: Yes" if duplicates else "Duplicate Accounts Found: No"
    
    if duplicates:
        print("Duplicate accounts:", ", ".join(duplicates))
    return duplicate_status

usernames = ["misterius_life", "sravani_1", "misterio_vida", "sravani_1", "misterio_vida"]
detect_duplicate_usernames(usernames)
```

```python
# 5.HEALTHCARE -- APPOINTMENT ELIGIBILITY CHECKER

def check_eligibility(patient_age):

    if patient_age >= 18:
        return "Eligible"
    else:
        return "Not Eligible"

patient_age = 21
eligibility_status = check_eligibility(patient_age)
print(f"Patient Age: {patient_age}")
print(f"Eligibility Status: {eligibility_status}")
```

```python
# 6.AGRICULTURE -- PREMIUM CROP PRICE FILTER

def premium_crop_filter(prices):
    premium_crops = [price for price in prices if price > 2000]
    return premium_crops

crop_prices = [1500, 2500, 1800, 3200]
premium_crops = premium_crop_filter(crop_prices)
print("Premium Crops:", premium_crops)
```

```python
# 7.SYSTEM MONITORING -- APPLICATION HEALTH CHECKER

def application_health_checker(error_count):
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    
    return f"Error Count: {error_count}\nSystem Status: {status}"

error_count = 7
print(application_health_checker(error_count))
```

```python
# 8.BANKING -- DAILY TRANSACTION LIMIT CHECKER

def daily_transaction_limit_checker(amount):
    DAILY_LIMIT = 50000
    if amount <= DAILY_LIMIT:
        return "Approved"
    else:
        return "Rejected"

transaction_amount = 60000
transaction_status = daily_transaction_limit_checker(transaction_amount)
print(f"Transaction Amount: {transaction_amount}")
print(f"Transaction Status: {transaction_status}")
```

```python
# 9.E-LEARNING -- STUDENT ATTENDENCE ELIGIBILITY SYSTEM

def check_attendance_eligibility(total_classes, attended_classes):
    attendance_percentage = (attended_classes / total_classes) * 100
    if attendance_percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"
    return attendance_percentage, eligibility

students = [
    {"name": "Student1", "total_classes": 100, "attended_classes": 80},
    {"name": "Student2", "total_classes": 100, "attended_classes": 70},
]

for student in students:
    attendance_percentage, eligibility = check_attendance_eligibility(student["total_classes"], student["attended_classes"])
    print(f"Attendance Percentage: {attendance_percentage:.1f}")
    print(f"Exam Eligibility: {eligibility}")
```

```python
# 10.SMART ELECTRICITY BILL ANALYZER

def electricity_bill(units):
    # Calculate bill based on slabs
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = 100 * 3 + (units - 100) * 5
    else:
        bill = 100 * 3 + 100 * 5 + (units - 200) * 7
    
    if bill < 500:
        usage_status = "Low Usage"
    elif bill <= 1500:
        usage_status = "Moderate Usage"
    else:
        usage_status = "High Usage"
    
    return bill, usage_status

units = int(input("Enter units consumed: "))
bill, status = electricity_bill(units)
print(f"Total electricity bill: ₹{bill}")
print(f"Usage classification: {status}")
```
