## 1 . SMART LIGHT CONTROLLER

```python
class SmartLight:
    def __init__(self, name):
        self.name=name
        self.status=False
    def turn_on(self):
        self.status=True
        print(f"{self.name} is ON")
    def turn_off(self):
        self.status=False
        print(f"{self.name} is OFF")
    def get_status(self):
        return self.status

light=SmartLight("Bedroom Light")
light.turn_on()
light.get_status()
```

## 2.EMPLOYEE ID CARD SYSTEM

```python
class Employee:
    def __init__(self, name, emp_id, department):
        self.name=name
        self.emp_id=emp_id
        self.department=department
    def display_id_card(self):
        print("Employee ID Card")
        print(f"Name : {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")

name=input("Enter Employee Name:")
emp_id=input("Enter Employee ID:")
department=input("Enter Department:")

employee=Employee(name, emp_id, department)
employee.display_id_card()
```

## 3.MOBILE CONTACT RECORD

```python
class Contact:
    def __init__(self, name, phone):
        self.name=name
        self.phone=phone
    def display_contact(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")

name=input("Enter the Name:")
number=input("Enter the Number:")

contact=Contact(name,number)
contact.display_contact()
```

## 4.PRODUCT PRICE TAG GENERATOR

```python
class ProductPriceTag:
    def __init__(self, product_name, price):
        self.product_name=product_name
        self.price=price

    def price_tag(self):
        print(f"Product: {self.product_name}")
        print(f"Price: {self.price}")

product=ProductPriceTag("Headphones",2499)
product.price_tag()
```

## 5.MOVIE RATING DISPLAY SYSTEM

```python
class MovieRating:
    def __init__(self,movie_name, rating):
        self.movie_name=movie_name
        self.rating=rating
    def details(self):
        print(f"Movie: {self.movie_name}")
        print(f"Rating: {self.rating}")

movie=MovieRating("Inception", 4.8)
movie.details()
```

## 6.DELIVERY ADDRESS MANAGER

```python
class DeliveryAddressManager:
    def __init__(self, customer_name, address):
        self.customer_name=customer_name
        self.address=address
    def delivery_details(self):
        print("Delivery Details")
        print(f"Customer: {self.customer_name}")
        print(f"Address: {self.address}")

delivery=DeliveryAddressManager("Uma","Kodad")
delivery.delivery_details()
```
