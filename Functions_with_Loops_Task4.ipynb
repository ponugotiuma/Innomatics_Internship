```python
# 1.MOBILE RECHARGE VALIDATION SYSTEM

def validate_recharge(amount):
    valid_plans=[199,299,399,599]
    if amount>=50 and amount in valid_plans:
        return True
    return False

def mobile_recharge_system():
    while True:
        try:
            amount=int(input("Enter recharge amount:"))
            if validate_recharge(amount):
                print("Recharge successful!")
                break
            else:
                print("Invalid amount. Valid plans are 199,299,399,599 and amount must be >= 50.Try again.")
        except ValueError:
            print("Please enter a numberic value.")

mobile_recharge_system()
```

```python
# 2. INVENTORY REORDER ALERT SYSTEM

def inventory_reorder_alert(inventory_dict):
    for product, stock in inventory_dict.items():
        if stock<15:
            print(f"{product}: Reorder Alert")
        else:
            print(f"{product}: Stock OK")

inventory={
    "Product A":10,
    "Product B":20,
    "Product C":5
}

inventory_reorder_alert(inventory)
```

```python
# 3.STUDENT RESULT PROCESSING SYSTEM

def process_student_result(marks):
    total=sum(marks)
    average=total/len(marks)
    result="Pass" if average>=50 else "Fail"
    print(result)
    return average,result

student_marks=[45,55,60,40]
avg,res=process_student_result(student_marks)
print(f"Average Marks: {avg}")
```

```python
# 4.CAB FARE ESTIMATOR WITH RETRY OPTION

def calculate_fare(distance, is_peak_hour):
    base_fare = 50
    per_km_charge = 12
    fare = base_fare + (distance * per_km_charge)
    if is_peak_hour:
        fare *= 1.25  # 25% extra for peak hour
    return fare

def cab_fare_calculator():
    retry = 'y'
    while retry.lower() == 'y':
        try:
            distance = float(input("Enter the distance traveled (in km): "))
            peak_hour = input("Is it peak hour? (y/n): ")
            is_peak = peak_hour.lower() == 'y'
            
            fare = calculate_fare(distance, is_peak)
            print(f"The total cab fare is: ₹{fare:.2f}")
            
            retry = input("Do you want to calculate again? (y/n): ")
        except ValueError:
            print("Invalid input. Please enter numeric values for distance.")
            retry = input("Do you want to try again? (y/n): ")


cab_fare_calculator()
```

```python
# 5.EMPLOYEE ATTENDENCE ELIGIBILITY

def attendence_eligibility(P,A):
    present_days=sum(1 for day in A if day=='P')
    attendence_percentage=(present_days/P)*100

    if attendence_percentage>=75:
        return "Eligible"
    else:
        return "Not Eligible"

total_days=10
attendence_list=['P','P','A','P','P','P','A','P','P','P']
print(attendence_eligibility(total_days, attendence_list))
```

```python
# 6.PASSWORD STRENGTH CHECKER

def check_password_strength(password):
    
    if len(password) < 8:
        return False
    
    has_digit = False
    has_special = False
    
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in ['@', '#', '$']:
            has_special = True
    
    if not has_digit or not has_special:
        return False
    return True

password = input("Enter a password: ")
if check_password_strength(password):
    print("Password is strong.")
else:
    print("Password is weak. It must be at least 8 characters long and contain at least one digit and one special character (@, #, $).")
```
