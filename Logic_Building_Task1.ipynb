```python
# 2.PASS / FAIL ANALYZER

# Given marks list
marks = [45, 78, 90, 33, 60]

# Initializing for Counting
pass_count = 0
fail_count = 0

# For result
for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("Total Students:", len(marks))
print("Total Passed:", pass_count)
print("Total Failed:", fail_count)

```

```python
# 3.SIMPLE DATA CLEANER

# Given list
names = [" Alice ", "bob", " CHARLIE "]

# A Clean Dataset for Result(To Store)
cleaned_names = []

for name in names:
    cleaned_name = name.strip().lower()
    cleaned_names.append(cleaned_name)

# Final Cleaned list
print("Cleaned Names:", cleaned_names)
```

```python
# 4.MESSAGE LENGTH ANALYZER

# Given messages
messages = ["Hi", "Welcome to the platform", "OK"]

# Analyzing messages sizes
for message in messages:
    length = len(message)
    print("Message:", message)
    print("Length:", length)
    
    if length > 10:
        print("Message longer than 10 characters")
    
    print()  
```

```python
# 5.ERROR MESSAGE DETECTOR

# Given System logs
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

# Counting ERROR entries
error_count = 0

for log in logs:
    if log == "ERROR":
        error_count += 1

# Total Error Count
print("Total ERROR entries:", error_count)
```
