```python
# 1.EMPLOYEE PERFORMANCE BONUS ELIGIBILITY

def find_top_performers(employees):
    #Find the maximum performance score
    max_score=max(employees.values())

    #Identify all employees with the highest score
    top_performers=[name for name, score in employees.items() if score==max_score]
    return top_performers, max_score

# Initializing dictionary
employees={
    "Ravi":92,
    "Anita":88,
    "Kiran":92,
    "Suresh":85
}

#Top Performers and their Score
top_performers, max_score= find_top_performers(employees)

print(f"Top Performers Eligible for Bonus:{','.join(top_performers)} (Score:{max_score})")
```

```python
# 2.SEARCH QUERY KEYWORD ANALYSIS

def search_query_keyword_analysis(query):
    #Converting input to Lowercase
    query=query.lower()

    #Ignore common punctuation( replace with Spaces)
    query=''.join(char if char.isalnum() or char.isspace() else ' ' for char in query)

    #Split query into words
    words=query.split()

    #Count frequency of each keyword
    word_count={}
    for word in words:
        if word in word_count:
            word_count[word]+=1
        else: 
            word_count[word]=1
    #Filtering keywords searched more than once
    result={word:count for word, count in word_count.items() if count>1}
    return result

input_query="Buy mobile phone buy mobile online"
output=search_query_keyword_analysis(input_query)
print(output)
```

```python
# 3.SENSOR DATA VALIDATION

def validate_sensor_data(sensor_readings):
    valid_readings=[(hour,reading) for hour,reading in enumerate(sensor_readings) if reading%2==0]
    return valid_readings

sensor_readings=[3,4,7,8,10,12,5]
output=validate_sensor_data(sensor_readings)
print(output)
```

```python
# 4.EMAIL DOMIAIN USAGE ANALYSIS 

def email_domain_analysis(emails):
    domain_count={}
    total_emails=len(emails)

    for email in emails:
        domain=email.split('@')[1]
        domain_count[domain]=domain_count.get(domain,0)+1

    result={domain:(count/total_emails)*100 for domain,count in domain_count.items()}
    return result

emails=[
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

output=email_domain_analysis(emails)
for domain,percentage in output.items():
    print(f"{domain}:{percentage:.0f}%")
```

```python
# 5.SALES SPIKE DETECTION

def detect_sales_spike(sales):
    avg_sales=sum(sales)/len(sales)

    spike_threshold=avg_sales*1.3
    spikes=[(i+1, sale) for i,sale in enumerate(sales) if sale>spike_threshold]

    for day,sale in spikes:
        print(f"Day {day}: {sale}")

sales=[1200,1500,900,2200,1400,3000]

detect_sales_spike(sales)
```

```python
# 6.DUPLICATE USER ID DETECTION

def detect_duplicate_user_ids(user_ids):
    # Create a dictionary to count occurrences of each ID
    id_count = {}
    for user_id in user_ids:
        if user_id in id_count:
            id_count[user_id] += 1
        else:
            id_count[user_id] = 1
    
    # Filter IDs that appear more than once and display frequency
    duplicates = {uid: count for uid, count in id_count.items() if count > 1}
    return duplicates

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

duplicate_ids = detect_duplicate_user_ids(user_ids)
for user, times in duplicate_ids.items():
    print(f"{user} â†’ {times} times")
```
