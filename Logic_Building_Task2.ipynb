```python
#1.UNIQUE WORDS IN A SENTENCE

# Input sentence
sentence = input("Enter sentence: ")

# Split sentence into words
words = sentence.split()

# Convert to set to get unique words
unique_words = set(words)

print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

```python
#2.HIGHEST SALARY FROM THE EMPLOYEE DATA

# Input Dictionary
employees = {
    "Swathi": 75000,
    "Anita": 68000,
    "Charan": 72000
}

# For Highest Salary A Variable
employee=""
highest=0

for name in employees:
    if employees[name] > highest:
        highest = employees[name]
        employee = name
print(f"The Highest Salary is : {employee} - {highest}")
```

```python
#3.MAXIMUM AND MINIMUM VALUES

# Initializing List
list = [45, 22, 89, 10, 66]

# Initializing Max and Min Variables
maximum=list[0]
minimum=list[0]

for num in list:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num
        
print("List:", list)
print("Max:", maximum)
print("Min:", minimum)
```

```python
# 4.COUNT PRODUCTS ABOVE A PRICE THHRESHOLD

# Initializing List 0f Prices
prices = [450, 1200, 1100, 1500, 300]

# Counter for products above 1000
count = 0

for price in prices:
    if price > 1000:
        count += 1

print("Products above 1000:", count)

```

```python
# 5.CALCULATE ATTENDENCE PERCENTAGE

# Attendance list
attendance = ["P", "A", "A", "P", "P"]

# Count present days
present_count = 0

for day in attendance:
    if day == "P":
        present_count += 1

# Calculate percentage
total_days = len(attendance)
percentage = (present_count / total_days) * 100

print("Attendance Percentage:", percentage)
```

```python
# 6.REMOVE DUPLICATE PHONE NUMBERS

# Given list of phone numbers
phone_numbers = [9876543210, 9123456789, 9876543210]

# Convert list to set to remove duplicates
unique_numbers = set(phone_numbers)

print("Unique phone numbers:", unique_numbers)
```

```python
# 7.COUNT CHARACTER FREQUENCY

# Initializing string
text = "pythonp"

# Dictionary to store counts
char_count = {}

# Count each character
for ch in text:
    if ch in char_count:
        char_count[ch] += 1
    else:
        char_count[ch] = 1

print(char_count)
```

```python
# 8.CONVERT LIST TO TUPLE

# A list
numbers = [10, 20, 30]

# Convert list to tuple
to_tuple = tuple(numbers)

print("Tuple:", to_tuple)
```

```python
# 9.CHECK IF A KEY EXISTS IN A DICTIONARY

# The Problem-2 dictionary
employees = {
    "Swathi": 75000,
    "Anita": 68000,
    "Charan": 72000
}

# Key to check
key = "Charan"

if key in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```

```python
# 10.CALCULATE AVERAGE MARKS

# The marks list
marks = [70, 75, 80, 65]

# Calculating average
total = sum(marks)
count = len(marks)
average = total / count

print("Average Marks:", average)
```
